```python
class MyError(Exception):
    pass
```

```python
year = int(input("Enter a year: "))
age = 2024 - year
try:    if age < 0:
        raise MyError("You haven't been born yet!")
    elif age > 150:
        raise MyError("That's an unrealistic age!")
    else:
        print(f"You are {age} years old.")
except MyError as e:
    print(f"Error: {e}")
```
