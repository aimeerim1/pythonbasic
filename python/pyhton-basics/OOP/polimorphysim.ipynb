```python
### Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclasses must implement this method")  
class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"
class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"
# Example usage
if __name__ == "__main__":
    dog = Dog("Buddy")
    cat = Cat("Whiskers")
    print(dog.speak())  # Output: Buddy says Woof!
    print(cat.speak())  # Output: Whiskers says Meow!    

    #Function that demonstrates polymorphism
    def animal_speak(animal):
        print(animal.speak())
    animal_speak(dog)  # Output: Buddy says Woof!
    animal_speak(cat)  # Output: Whiskers says Meow!
```

```python
#Polymorphism with functions and methods
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement this method")
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
# Example usage
if __name__ == "__main__":
    circle = Circle(5)
    rectangle = Rectangle(4, 6)
    print(f"Circle area: {circle.area()}")  # Output: Circle area: 78.5
    print(f"Rectangle area: {rectangle.area()}")  # Output: Rectangle area: 24

    # Function that demonstrates polymorphism
    def print_area(shape):
        print(f"The area is: {shape.area()}")
    print_area(circle)  # Output: The area is: 78.5
    print_area(rectangle)  # Output: The area is: 24
```
